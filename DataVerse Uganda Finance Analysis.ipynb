```python
int_fin_dataset = ['Family_Size']
float_fin_dataset = ['Distance_to_Bank_km', 'Financial_Stress_Score']
obj_fin_dataset = ['Household_ID']
cat_fin_dataset = ['Location', 'Education_Level', 'Employment_Type', 'Has_Savings_Account', 'Uses_Savings_Group', 'Uses_Mobile_Money', 'Borrowing','Selling_Assets', 'Reducing_Spending']

for col in fin_dataset.columns:
    if col in int_fin_dataset:
        fin_dataset[col] = fin_dataset[col].astype('int64')
    elif col in float_fin_dataset:
        fin_dataset[col] = fin_dataset[col].astype('float64')
    elif col in obj_fin_dataset:
        fin_dataset[col] = fin_dataset[col].astype('object')
    elif col in cat_fin_dataset:
        fin_dataset[col] = fin_dataset[col].astype('category')
```

```python
sns.set_style('whitegrid')
sns.distplot(fin_dataset['Financial_Stress_Score'], kde=True, color='red', bins=40)
```

```python
fin_dataset['Financial_Stress_Score'].describe()
```

```python
Q1 = 35.565000
Q3 = 45.050000

IQR = Q3 - Q1

lower_outlier = Q1 - 1.5*IQR
print(lower_outlier)
```

```python
# With the outlier on the Financial_Stress_Score, we can group households into 2 classes of Most extremely stressed and Leaset extremely stresed

most_financial_stressed = fin_dataset[fin_dataset['Financial_Stress_Score'] >= lower_outlier]
least_financial_stressed = fin_dataset[fin_dataset['Financial_Stress_Score'] < lower_outlier]
```

### Coping Mechanism during Financial Stress using the Financial_Stress_Score

```python
most_financial_stressed
```

```python
# Now, we proceed to investigate how they cope during down season of low income

Borrowing = np.array(most_financial_stressed.groupby(['Borrowing'])['Household_ID'].count())
Selling_Assets = np.array(most_financial_stressed.groupby(['Selling_Assets'])['Household_ID'].count())
Reducing_Spending = np.array(most_financial_stressed.groupby(['Reducing_Spending'])['Household_ID'].count())

most_financial_stressed_df = pd.DataFrame({
                                            'Borrowing':Borrowing,
                                            'Selling_Assests':Selling_Assets,
                                             'Reducing_Spending':Reducing_Spending
}).reset_index().rename(columns={'index':'Index'})

most_financial_stressed_df
```

```python
# Creating a melted or transposes dataframe to aid visualization
df_melted = most_financial_stressed_df.melt(id_vars='Index',
                                            var_name='Feature',
                                            value_name='Value')

most_financial_stressed_df['Index'] = most_financial_stressed_df['Index'].astype('str')

plt.figure(figsize=(10,6))
sns.barplot(x='Feature', y='Value', hue='Index', data=df_melted)
plt.title('Coping Mechanism of Most_Financial_Stressed household')
plt.xlabel('Coping Strategy')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Feature Engineering

```python
total_income = pd.DataFrame(season_dataset.groupby(['Household_ID']).mean()['Total_Income']).reset_index()
total_income
```

```python
# Adding the mean of the Total_Income of each households from the Season dataset to the Financial Dataset

fin_dataset = fin_dataset.merge(total_income, on='Household_ID')
fin_dataset
```

### Feature Endoing

```python
fin_dataset.info()
```

```python
fin_dataset[['Location', 'Education_Level', 'Employment_Type']]
```

```python
cat_features = fin_dataset[['Location', 'Education_Level', 'Employment_Type']]

#Converting Categorical features

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first')

encoded_data = encoder.fit_transform(cat_features)

encoded_columns = encoder.get_feature_names_out()

encoded_df = pd.DataFrame(encoded_data,columns=encoded_columns,index=fin_dataset.index)

trans_dataset = pd.concat([fin_dataset, encoded_df], axis=1)    #Renaming the new transformed dataset
```

```python
trans_fin_dataset = trans_dataset.drop(cat_features, axis=1)
```

```python
trans_fin_dataset
```

```python
#Reanaming columns to avoid gaps in column names during deployment

trans_fin_dataset = trans_fin_dataset.rename(columns={
                                   'Location_Fort Portal':'Location_Fort_Portal',
                                    'Employment_Type_Livestock Rearing':'Employment_Type_Livestock_Rearing',
                                    'Employment_Type_Small Business':'Employment_Type_Small_Business',
                                    'Employment_Type_Subsistence Farming':'Employment_Type_Subsistence_Farming' 
})
```

### X/y Split

```python
# Spliting data to features and target variables
X = trans_fin_dataset.drop(['Financial_Stress_Score', 'Household_ID'], axis=1).values
y = trans_fin_dataset['Financial_Stress_Score'].values
```

### Train/Test Split

```python
from sklearn.model_selection import train_test_split

# Spliting data to test and train set
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.80, random_state=10)
```

### Model Training: LinearRegression

```python
from sklearn.linear_model import LinearRegression

# Define the model
regressor = LinearRegression()

regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
# Feature Importance

columns = X = trans_fin_dataset.drop(['Financial_Stress_Score', 'Household_ID'], axis=1).columns
feature_importance = pd.DataFrame({'Feature':columns,
                                  'Importance':regressor.coef_},
                                 index=np.arange(len(columns)))

#feature_importance['Percentage'] = (feature_importance['Importance']/feature_importance['Importance'].sum())*100
feature_importance = feature_importance.sort_values('Importance')
feature_importance
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(y=feature_importance['Feature'], x=feature_importance['Importance'], palette="coolwarm")
plt.xlabel("Coefficient", fontsize=12)
plt.title('Relevance of the Features', fontsize=14)
plt.grid()
plt.tight_layout()
```

### Model Training: OLS rRegression

```python
import statsmodels.api as sm

y_train = np.array(y_train, dtype=np.float64)

X_train_df = pd.DataFrame(X_train)
column_name = list(trans_fin_dataset.drop(['Financial_Stress_Score', 'Household_ID'],axis=1).columns)
X_train_df.columns = column_name
X_train = np.array(X_train, dtype=np.float64)



lm1 = sm.OLS(y_train,sm.add_constant(X_train)).fit()
print(lm1.summary(xname=['const']+column_name))
```

```python
# Feature Importance

columns = X = trans_fin_dataset.drop(['Financial_Stress_Score', 'Household_ID'], axis=1).columns
feature_importance1 = pd.DataFrame({'Feature':columns.insert(0,'constant'),
                                  'Importance':lm1.params},
                                 index=np.arange(len(columns.insert(0,'constant'))))

#feature_importance['Percentage'] = (feature_importance['Importance']/feature_importance['Importance'].sum())*100
feature_importance1 = feature_importance1[1:].sort_values('Importance')
feature_importance1
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(y=feature_importance1['Feature'], x=feature_importance1['Importance'], palette="coolwarm")
plt.xlabel("Coefficient", fontsize=12)
plt.title('Relevance of the Features', fontsize=14)
plt.grid()
plt.tight_layout()
```
